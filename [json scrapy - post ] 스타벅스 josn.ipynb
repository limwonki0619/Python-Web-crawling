## **스타벅스 매장 정보 크롤링**

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
SB_data = {
    'ins_lat':'37.56682', # 지정한 위도와 경도에서 가까운 순으로 나열
    'ins_lng':'126.97865',
    'p_sido_cd':'', # 01=서울시, 08=경기 ... 16=제주
    'p_gugun_cd':'',  # 세부지역 (지정하지 않으면 시/도 전체)
    'in_biz_cd':'',
    'set_date':'',
    'iend':'10000',
}
    
SB_url = 'https://www.istarbucks.co.kr/store/getStore.do'
SB_r = requests.post(SB_url, data=SB_data)

print(SB_r.text[:1000], sep='\t') # 수신된 데이터의 앞부분만 확인
```

```python
SB_jo = json.loads(SB_r.text)
SB_df_raw = json_normalize(SB_jo, 'list')
SB_df = SB_df_raw.copy()
```
